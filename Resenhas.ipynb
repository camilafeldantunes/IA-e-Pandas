```python
from groq import Groq

client = Groq()

def resposta_perguntas(coluna_reviews):
    lista_respostas = []
    lista_sentimentos = []

    for i, resenha in enumerate(coluna_reviews):

        completion = client.chat.completions.create(
            model="openai/gpt-oss-20b",
            messages=[
                {
                    "role": "user",
                    "content": resenha + ". Responda essa resenha com: Neutro, Positivo ou Negativo"
                },
            ],
            temperature=1,
            max_completion_tokens=8192,
            top_p=1,
            reasoning_effort="medium",
            stream=True
        )


        resposta_texto = ""

        for chunk in completion:
            conteudo = chunk.choices[0].delta.content or ""

            resposta_texto += conteudo



        lista_respostas.append({
            "Resenha": resenha,
            "Resposta": resposta_texto
        })
        lista_sentimentos.append(resposta_texto)


    return lista_sentimentos

```

```python
resposta_perguntas(coluna_reviews)
```

```python
df_reviews["sentimentos"] = resposta_perguntas(coluna_reviews)
```
