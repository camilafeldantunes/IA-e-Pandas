```python
lista_de_perguntas = [
    "Qual a diferença entre atividade física e exercício físico?",
    "Quais são os principais benefícios da prática regular de esportes para a saúde mental?",
    "Por que o aquecimento é tão importante antes de praticar uma atividade física?",
    "Como o esporte pode contribuir para o desenvolvimento social e emocional de uma pessoa",
    "Quais são os riscos do sedentarismo a longo prazo?",
    "Qual a importância do alongamento após a prática esportiva?",
    "Como a alimentação influencia o desempenho esportivo?",
    "Quais esportes são mais indicados para melhorar o condicionamento cardiovascular?",
    "De que forma a prática de atividade física pode ajudar na prevenção de doenças crônicas?",
    "Como escolher uma atividade física adequada de acordo com a idade e condição física?"
]
```

```python
with open("lista-perguntas.txt", "w", encoding="utf-8") as arquivo:
  for pergunta in lista_de_perguntas:
    arquivo.write(pergunta + "\n")
```

```python
nova_lista_de_perguntas = []

with open("lista-perguntas.txt", "r", encoding="utf-8") as arquivo:
  for linha in arquivo:
    nova_lista_de_perguntas.append(linha.strip())
```

```python
import os
from google.colab import userdata
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')
```

```python
from groq import Groq
import pandas as pd

client = Groq()

def resposta_perguntas(nova_lista_de_perguntas):
    lista_respostas = []

    for i, elemento in enumerate(nova_lista_de_perguntas):

        completion = client.chat.completions.create(
            model="openai/gpt-oss-20b",
            messages=[
                {
                    "role": "user",
                    "content": elemento + ". Responda essa pergunta de forma clara e fácil de entender, em no máximo 2 parágrafos"
                },
            ],
            temperature=1,
            max_completion_tokens=8192,
            top_p=1,
            reasoning_effort="medium",
            stream=True
        )

        print(f"Pergunta {i+1}:")

        resposta_texto = ""

        for chunk in completion:
            conteudo = chunk.choices[0].delta.content or ""
            print(conteudo, end="")
            resposta_texto += conteudo

        print("\n" + "-"*50 + "\n")

        lista_respostas.append({
            "Pergunta": elemento,
            "Resposta": resposta_texto
        })

    # Criando DataFrame
    df = pd.DataFrame(lista_respostas)

    # Salvando CSV
    df.to_csv("respostas.csv", index=False, encoding="utf-8")

    return df


```

```python
resposta_perguntas(nova_lista_de_perguntas)
```
