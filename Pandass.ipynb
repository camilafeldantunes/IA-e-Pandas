Manipulando dados com Pandas

```python
import pandas as pd
import numpy as np

# Gerar 50 linhas de dados para as colunas especificadas
num_rows = 50

product_names = [f'Produto_{i}' for i in range(1, num_rows + 1)]
categories = np.random.choice(['Eletrônicos', 'Roupas', 'Livros', 'Alimentos', 'Decoração'], num_rows)
prices = np.round(np.random.uniform(10.00, 1000.00, num_rows), 2)
items_sold = np.random.randint(5, 500, num_rows)
ratings = np.round(np.random.uniform(1.0, 5.0, num_rows), 1)

data = {
    'nome do produto': product_names,
    'categoria do produto': categories,
    'Preço': prices,
    'quantos itens do produto foram vendidos': items_sold,
    'avaliação do produto': ratings
}

df_produtos = pd.DataFrame(data)

display(df_produtos.head())
```
